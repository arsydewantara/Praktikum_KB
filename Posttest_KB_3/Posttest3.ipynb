```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HRDataset_v14.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

# Bar Plot

```python
x = list(df["PerformanceScore"])
y = list(df["Salary"])


plt.figure(figsize=(20,15))
plt.bar(x,y,color='red')
plt.xlabel('Skor Performa')
plt.ylabel('Gaji')
plt.title("Perbandingan Gaji berdasarkan Skor Performa")
```

#### Kesimpulan
Dari Grafik diatas bisa dilihat perbandingan Gaji berdasarkan 4 kategori performa skor, dari 4 kategori performa skor, pegawai dengan performa memenuhi atau Fully Meets memiliki gaji yang sangat tinggi, pegawai dengan performa melebihi atau Exceeds memiliki gaji tinggi setelah Fully Meets, kemudian pegawai dengan performa Needs improvement atau butuh peningkatan memiliki gaji yang tidak terlalu tinggi dan terakhir pegawai dengan performa PIP memiliki gaji yang cukup Rendah

# Pie Chart

```python
df["PerformanceScore"].value_counts()
```

```python
data_pie = df["PerformanceScore"].value_counts().rename_axis("PerformanceScore").reset_index(name="id_counts")
```

```python
data_pie.head(10)
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie["id_counts"],labels=data_pie['PerformanceScore'],startangle=90,autopct='%.1f%%' ,explode=(0,0,0,0))
plt.title('Persentase Perbandingan Skor Performa')
plt.show()
```

#### Kesimpulan
dari grafik diatas kita bisa melihat persentase skor performa pegawai. Pegawai dengan performa Memenuhi atau Fully Meets terbilang cukup banyak yaitu sebesar 78.1%, Pegawai dengan performa melebihi atau Exceeds sebesar 11.9%, pegawai dengan performa Needs Improvement atau butuh peningkatan terbilang cukup sedikit yaitu sebesar 5.8%, dan pegawai dengan performa yang rendah sebesar 4.2%

# Scatter Plot

```python
sns.scatterplot(data=df,x ="PerformanceScore",y= 'SpecialProjectsCount')
plt.title('Perbandingan Skor Performa dan Jumlah Proyek Spesial')
plt.xlabel('Skor Performa')
plt.ylabel('Jumlah Proyek Spesial')
plt.show()
```

#### Kesimpulan
dari grafik diatas bisa kita lihat Pegawai dengan performa Fully Meets memiliki proyek spesial yang banyak, pegawai dengan performa Exceeds memiliki proyek spesial terbanyak kedua setelah Fully Meets, pegawai dengan performa Needs Improvement memiliki memiliki jumlah proyek spesial yang sama dengan pegawai dengan performa PIP
